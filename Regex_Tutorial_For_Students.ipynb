## 1. What is a Regular Expression (Regex)?

- A **regular expression** is a special string pattern used to **search**, **match**, and **manipulate** text.
- Instead of manually checking each character, regex lets us describe what we are looking for using a **pattern language**.

Example real-world use cases:
- Check if a password is strong.
- Validate email or mobile number.
- Find all hashtags in a tweet.
- Replace multiple spaces with a single space.
- Extract dates, amounts, names from text.


```

import re

# Always start with importing the 're' module in Python
```


## 2. Basic Regex Concepts

Before going deep, understand basic ideas:

- **Literal characters**: Match exactly the same character.
  - Pattern: `"cat"` → matches the substring `"cat"` in `"my cat is cute"`
- **Case-sensitive** by default.
- Regex is applied on strings.

Let's see a simple example.


```

text = "My cat is sleeping. I love my cat."
pattern = "cat"

matches = re.findall(pattern, text)
print("Matches:", matches)
print("Number of times 'cat' appears:", len(matches))
```


## 3. Character Classes and Ranges

Character classes are defined using **square brackets `[]`**.

- `[abc]` → match **a OR b OR c**
- `[0-9]` → match any digit from 0 to 9
- `[a-z]` → match any lowercase letter
- `[A-Z]` → match any uppercase letter
- `[a-zA-Z]` → any letter (upper or lower)
- `[^0-9]` → `^` inside `[]` means **NOT**, so this means any **non-digit**

Let's see examples.


```

text = "Room 101, Floor 2, Block B"

# Find all digits
digits = re.findall(r"[0-9]", text)
print("Digits:", digits)

# Find all letters
letters = re.findall(r"[A-Za-z]", text)
print("Letters:", letters)

# Find all non-digits
non_digits = re.findall(r"[^0-9]", text)
print("Non-digits (first 30):", non_digits[:30])
```


## 4. Quantifiers

Quantifiers specify **how many times** a character or group can repeat:

- `*` → 0 or more times  
- `+` → 1 or more times  
- `?` → 0 or 1 time (optional)  
- `{m}` → exactly m times  
- `{m,}` → at least m times  
- `{m,n}` → between m and n times (inclusive)

Examples:
- `a*` → `""`, `a`, `aa`, `aaa`, …  
- `a+` → `a`, `aa`, `aaa`, … (but **not** empty)  
- `[0-9]{3}` → exactly 3 digits

Let's try with Python.


```

text = "aaa a aa aaaa aaaaa"

# One or more 'a'
print("Pattern: a+")
print(re.findall(r"a+", text))

text2 = "Version: v1 v12 v123 v1234"

# 1 to 3 digits after 'v'
print("\nPattern: v[0-9]{1,3}")
print(re.findall(r"v[0-9]{1,3}", text2))
```


## 5. Anchors: `^` and `$`

Anchors do not match characters; they match **positions**:

- `^` → start of the string
- `$` → end of the string

Examples:
- `^Hello` → matches `"Hello"` only if it is at the **start** of the string.
- `world$` → matches `"world"` only if it is at the **end** of the string.

This is useful for **validation**, e.g., check if entire string is only digits.


```

text1 = "Hello world"
text2 = "Well, Hello world again"

# Check if string starts with 'Hello'
print("Starts with Hello (text1):", bool(re.search(r"^Hello", text1)))
print("Starts with Hello (text2):", bool(re.search(r"^Hello", text2)))

# Check if string ends with 'world'
print("Ends with world (text1):", bool(re.search(r"world$", text1)))
print("Ends with world (text2):", bool(re.search(r"world$", text2)))
```


## 6. Grouping `()` and Alternation `|`

### Grouping with `()`
- Parentheses create **groups**.
- Useful for capturing parts of the match.

Example:  
- Pattern: `(\d{2})-(\d{2})-(\d{4})`  
  - Group 1: day  
  - Group 2: month  
  - Group 3: year  

### Alternation with `|`
- `a|b` → match `a` OR `b`.

Example:
- Pattern: `cat|dog` → matches `"cat"` or `"dog"`.


```

text = "My favourite pets are a cat, a dog, and a parrot."

# Alternation example
animals = re.findall(r"cat|dog", text)
print("Matched animals:", animals)

# Grouping example: date format DD-MM-YYYY
date_text = "Today's date is 05-12-2025."

match = re.search(r"(\d{2})-(\d{2})-(\d{4})", date_text)
if match:
    print("\nFull date:", match.group(0))
    print("Day:", match.group(1))
    print("Month:", match.group(2))
    print("Year:", match.group(3))
```


## 7. Special Sequences

Python regex provides some **shortcuts**:

- `\d` → digit (0–9), same as `[0-9]`
- `\D` → non-digit
- `\w` → word character: letters, digits, underscore (`[A-Za-z0-9_]`)
- `\W` → non-word character
- `\s` → whitespace (space, tab, newline)
- `\S` → non-whitespace

> Note: We use raw strings in Python by adding `r` before the string, e.g. `r"\d+"`.  
> This helps avoid problems with backslashes.


```

text = "User_123 logged in at 10:45 am."

digits = re.findall(r"\d+", text)
print("Digits (numbers):", digits)

words = re.findall(r"\w+", text)
print("Word-like tokens:", words)

spaces = re.findall(r"\s", text)
print("Whitespace characters count:", len(spaces))
```


## 8. Important `re` Functions

We will learn 5 important functions:

1. `re.search(pattern, text)` – Find **first** match **anywhere** in string.
2. `re.match(pattern, text)` – Check match **at the beginning** of string.
3. `re.findall(pattern, text)` – Return **list** of all non-overlapping matches.
4. `re.sub(pattern, replacement, text)` – Replace matches with something else.
5. `re.split(pattern, text)` – Split string based on regex pattern.

Let's see each with an example.


```

text = "Order ID: 4521, Price: 299.99 INR"

# 1. search
m = re.search(r"\d{4}", text)
print("search → first 4-digit number:", m.group(0))

# 2. match (checks only at start)
print("match at start:", re.match(r"Order", text) is not None)
print("match 'Price' at start:", re.match(r"Price", text) is not None)

# 3. findall
numbers = re.findall(r"\d+\.?\d*", text)
print("findall → all numbers:", numbers)

# 4. sub (replace digits with X)
masked = re.sub(r"\d", "X", text)
print("sub → masked digits:", masked)

# 5. split (split by comma and spaces)
parts = re.split(r",\s*", text)
print("split →", parts)
```


## 9. Practical Example 1: Simple Email Pattern

🧠 Goal: Check whether a string **looks like** an email.

A simple (not perfect) pattern:

- Username part: letters, digits, dot, underscore: `[A-Za-z0-9._]+`
- `@` symbol
- Domain part: letters, digits: `[A-Za-z0-9]+`
- Dot and extension: `\.[A-Za-z]{2,4}`

Full pattern:  
`^[A-Za-z0-9._]+@[A-Za-z0-9]+\.[A-Za-z]{2,4}$`

> `^` and `$` ensure that the whole string matches the pattern.


```

emails = [
    "student123@gmail.com",
    "teacher.mail@college.edu",
    "bad-email@@gmail.com",
    "noatsymbol.com",
    "abc@xy"
]

pattern = r"^[A-Za-z0-9._]+@[A-Za-z0-9]+\.[A-Za-z]{2,4}$"

for email in emails:
    if re.match(pattern, email):
        print(email, "→ VALID")
    else:
        print(email, "→ INVALID")
```


## 10. Practical Example 2: Simple Phone Number (Indian style)

🧠 Goal: Match 10-digit Indian phone numbers (basic version).

Assumptions:
- 10 digits
- May start with 6, 7, 8, or 9
- No country code handling here

Pattern:  
`^[6-9][0-9]{9}$`

Explanation:
- `^[6-9]` → first digit is 6, 7, 8, or 9
- `[0-9]{9}` → followed by exactly 9 digits
- `$` → end of string


```

phones = [
    "9876543210",
    "1234567890",
    "9999999999",
    "789654",
    "6789054321"
]

pattern = r"^[6-9][0-9]{9}$"

for ph in phones:
    print(ph, "→", "VALID" if re.match(pattern, ph) else "INVALID")
```


## 11. Practical Example 3: Cleaning Extra Spaces

Often in text, there are multiple spaces between words.  
We can use regex to replace **2 or more spaces** with **1 space**.

Pattern:  
- `\s+` → one or more whitespace characters


```

text = "This   sentence   has   too    many   spaces."

cleaned = re.sub(r"\s+", " ", text)
print("Original:", text)
print("Cleaned :", cleaned)
```
