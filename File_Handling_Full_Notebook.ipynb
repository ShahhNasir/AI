## 🔹 What is File Handling?

File handling allows Python programs to **store**, **read**, and **manipulate** text or data permanently.

### Why is it important?
- Save user data  
- Load configuration files  
- Process logs & documents  
- Work on datasets  


## 🔹 Opening Files in Python
We use the built-in `open()` function.

```
# Syntax:
# open(filename, mode)

f = open("sample.txt", "w")
f.write("Hello Students!")
f.close()
```


### 🔸 File Modes

| Mode | Meaning |
|------|---------|
| `r` | Read (file must exist) |
| `w` | Write (overwrite) |
| `a` | Append |
| `r+` | Read + Write |
| `w+` | Write + Read |
| `a+` | Append + Read |


## 🔹 Writing to Files

```
# Write example
with open("notes.txt", "w") as f:
    f.write("Python File Handling Basics")
    f.write("\nSecond line added.")
```

## 🔹 Reading Files

```
# Read entire file
with open("notes.txt", "r") as f:
    print(f.read())
```

### Reading line by line

```
with open("notes.txt", "r") as f:
    print(f.readline())
    print(f.readline())
```

### Read all lines using readlines()

```
with open("notes.txt", "r") as f:
    print(f.readlines())
```

## 🔹 Using `with open()` (Recommended)

Automatically closes file even if errors occur.

```
with open("data.txt", "w") as f:
    f.write("This is safe file handling!")
```

## 🔹 Appending to Files

```
with open("notes.txt", "a") as f:
    f.write("\nThis line is appended.")

with open("notes.txt", "r") as f:
    print(f.read())
```

## 🔹 Checking if a File Exists

```
import os
print(os.path.exists("notes.txt"))
print(os.path.exists("random.txt"))
```

## 🔹 Deleting Files

```
import os

if os.path.exists("delete_me.txt"):
    os.remove("delete_me.txt")
else:
    print("File not found.")
```

## 🔹 Working with Directories

```
import os

os.mkdir("myfolder")
print(os.listdir("."))  # List files/folders
os.rmdir("myfolder")
```

## 🟦 Mini Project 1: Save User Data to a File

```
name = "Nasir"
course = "Python"

with open("user_profile.txt", "w") as f:
    f.write(f"Name: {name}\nCourse: {course}")

print(open("user_profile.txt").read())
```

## 🟩 Mini Project 2: Word Counter

```
with open("story.txt", "w") as f:
    f.write("Python is powerful. Python is easy. Python is loved by students.")

with open("story.txt", "r") as f:
    words = f.read().split()

print("Total words:", len(words))
```

## 🟪 Mini Project 3: Copy File Content

```
with open("story.txt", "r") as src:
    content = src.read()

with open("story_backup.txt", "w") as dest:
    dest.write(content)

print("Copy Completed!")
```
