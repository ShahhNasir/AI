# ANN Showcase on MNIST (TensorFlow / Keras)
This notebook demonstrates **10 different ANN architectures** covering optimizers, activation functions, batch normalization, dropout, early stopping, regularization, and more.


```

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l1, l2
import numpy as np

# Load MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

print("Training shape:", x_train.shape)
```

```

def compile_and_train(model, optimizer='adam', epochs=5):
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    history = model.fit(x_train, y_train,
                        validation_split=0.2,
                        epochs=epochs,
                        batch_size=32,
                        verbose=1)
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print("Test Accuracy:", test_acc)
    return history
```

## 1️⃣ Basic ANN (ReLU + Adam)

```

model1 = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model1)
```

## 2️⃣ ANN with SGD Optimizer

```

model2 = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model2, optimizer='sgd')
```

## 3️⃣ ANN with Tanh Activation

```

model3 = keras.Sequential([
    layers.Dense(128, activation='tanh', input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model3)
```

## 4️⃣ ANN with LeakyReLU

```

model4 = keras.Sequential([
    layers.Dense(128, input_shape=(784,)),
    layers.LeakyReLU(),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model4)
```

## 5️⃣ ANN with Batch Normalization

```

model5 = keras.Sequential([
    layers.Dense(128, input_shape=(784,)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model5)
```

## 6️⃣ ANN with Dropout

```

model6 = keras.Sequential([
    layers.Dense(256, activation='relu', input_shape=(784,)),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model6)
```

## 7️⃣ ANN with Early Stopping

```

model7 = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])

model7.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

early_stop = EarlyStopping(monitor='val_loss', patience=2)

model7.fit(x_train, y_train,
           validation_split=0.2,
           epochs=20,
           batch_size=32,
           callbacks=[early_stop],
           verbose=1)

test_loss, test_acc = model7.evaluate(x_test, y_test, verbose=0)
print("Test Accuracy:", test_acc)
```

## 8️⃣ ANN with L2 Regularization

```

model8 = keras.Sequential([
    layers.Dense(128, activation='relu',
                 kernel_regularizer=l2(0.001),
                 input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model8)
```

## 9️⃣ Deep ANN (Multiple Hidden Layers)

```

model9 = keras.Sequential([
    layers.Dense(256, activation='relu', input_shape=(784,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model9)
```

## 🔟 ANN with Custom Learning Rate

```

optimizer = keras.optimizers.Adam(learning_rate=0.0001)

model10 = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(10, activation='softmax')
])
compile_and_train(model10, optimizer=optimizer)
```
